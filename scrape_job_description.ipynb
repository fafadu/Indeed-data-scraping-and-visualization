```python
import time
import random
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.keys import Keys
import undetected_chromedriver as uc




# Initialize the WebDriver
driver = uc.Chrome()
# Load job links from CSV
job_links = pd.read_csv("indeed_jobs.csv")["Job URL"].tolist()
job_details = []

# #Start visiting each link
for link in job_links:
    print(f"Accessing Indeed: {link}")
    driver.get(link)
    time.sleep(random.uniform(5, 8))  # Wait randomly, simulating real-person operation

    # Get company name
    try:
        time.sleep(random.uniform(2, 4))
        company_name = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, "div[data-testid='inlineHeader-companyName'] span a")
            )
        ).text
        print(f":company_name {company_name}")
       
    except Exception as e:
        
        company_name = "N/A"
        print(f":company_name {company_name}")
        
   # Get location:
    try:
        time.sleep(random.uniform(2, 4))
        location = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div[data-testid='inlineHeader-companyLocation']"))
        ).text
        print(f"location: {location}")
    except Exception as e:
        location = "N/A"
        print(f"location: {location}")

    # Get job description
    try:
        job_description = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "jobDescriptionText"))
        ).text
        
    except Exception as e:
        job_description = "N/A"    
       

    
    # Append the scraped details to the list
    job_details.append({
        "URL": link,
        "Company": company_name,
        "Location": location,
        "Job Description": job_description,
    })
    print("-" * 50)
    print(f"Successfully processed: {link}")


# Save to DataFrame
df = pd.DataFrame(job_details)

# Export as CSV
df.to_csv("job_details.csv", index=False)
print("Saved 'job_details.csv'")

# Close browser
driver.quit()
```
